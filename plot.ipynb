```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data.lastUpdated = pd.to_datetime(data.lastUpdated)
```

```python
BROWN_MUSHROOM = data[(data.item == "BROWN_MUSHROOM")]
```

```python
fig, ax = plt.subplots()
options = {"x":"lastUpdated", "kind":"line", "ax":ax, "figsize":(10,10)}
bp = BROWN_MUSHROOM.plot(**options, y= "buyPrice", color="red", label="buyPrice");
sp = BROWN_MUSHROOM.plot(**options, y = "sellPrice", color="green", label="sellPrice");
```
