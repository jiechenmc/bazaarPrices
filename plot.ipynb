```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv", index_col="lastUpdated")
```

```python
data.index = pd.DatetimeIndex(data.index)
```

```python
BROWN_MUSHROOM = data[(data.item == "BROWN_MUSHROOM")]
```

```python
fig, ax = plt.subplots()
bp = BROWN_MUSHROOM.reset_index().plot(x="lastUpdated", y="buyPrice", kind="scatter", color ='red', ax=ax, label="buyPrice");
sp = BROWN_MUSHROOM.reset_index().plot(x="lastUpdated", y="sellPrice", kind="scatter", color ='green', ax=ax, label="sellPrice");
```
